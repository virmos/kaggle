```python
df = pd.read_csv('data/2020/table_6.csv')
df.drop(columns=['Unnamed: 6','Unnamed: 7', 'Unnamed: 9','Unnamed: 10', 'Nng thn' ,'Thnh th' ], inplace=True)

new_header = df.iloc[0] 
df = df[1:] 
df.columns = new_header

# try to drop the first 6 rows
df = df[6:] 

df.head(3)
```

```python
# run once
df = df.loc[df['Trnh '] != "Tng s"]

df['N'] = df['N'].astype('float32')
df['Nam'] = df['Nam'].astype('float32')
df['N'] = df['N'].apply(lambda x : -x)

# L = np.array(df['Vng'].values.tolist()).ravel().tolist()
positions = pd.Series(df['Vng']).unique()
worktypes = pd.Series(df['Trnh ']).unique()
positions = np.array(['Trung du v min ni pha Bc', 'ng bng sng Hng',
       'Duyn hi min Trung', 'Ty Nguyn',
       'ng Nam B', 'ng bng sng Cu Long', 'H Ni', 'TP.HCM'])

data = df.groupby('Vng')['Nam', 'N'].apply(pd.Series.tolist).tolist()
data = np.array(data)
data.shape
```

```python
num_positions = data.shape[0]
num_worktypes = data.shape[1]
num_genders = data.shape[2]

width = 0.75
dimw = width / num_positions

fig, ax = plt.subplots(1,1, figsize=(20, 10))
x = np.arange(num_positions)
labels = positions

worktype_0_men = data[:, 0, 0]
worktype_1_men = data[:, 1, 0]
worktype_2_men = data[:, 2, 0]
worktype_3_men = data[:, 3, 0]
worktype_4_men = data[:, 4, 0]

worktype_0_women = data[:, 0, 1]
worktype_1_women = data[:, 1, 1]
worktype_2_women = data[:, 2, 1]
worktype_3_women = data[:, 3, 1]
worktype_4_women = data[:, 4, 1]

rects0 = ax.bar(x - width/6, worktype_0_men, dimw, label='Khng', color="blue")
rects1 = ax.bar(x - 2*width/6, worktype_1_men, dimw, label='Dy ngh', color="orange")
rects2 = ax.bar(x - 3*width/6, worktype_2_men, dimw, label='Trung cp',color="green")
rects3 = ax.bar(x, worktype_3_men, dimw, label='Cao ng',color="red")
rects4 = ax.bar(x + width/6, worktype_4_men, dimw, label='i hc ',color="brown")

ax.bar_label(rects0, padding=3)
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
ax.bar_label(rects4, padding=3)

rects0 = ax.bar(x - width/6, worktype_0_women, dimw, color="blue")
rects1 = ax.bar(x - 2*width/6, worktype_1_women, dimw, color="orange")
rects2 = ax.bar(x - 3*width/6, worktype_2_women, dimw, color="green")
rects3 = ax.bar(x + width/6, worktype_3_women, dimw, color="red")
rects4 = ax.bar(x + 2*width/6, worktype_4_women, dimw, color="brown")

ax.bar_label(rects0, padding=3)
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)
ax.bar_label(rects3, padding=3)
ax.bar_label(rects4, padding=3)

ax.axhline(y=0, color='black', linestyle='-')
ax.text(0, 60, "Nam", size=25, rotation=0,
         bbox=dict(boxstyle="round",
                   ec=(1., 0.5, 0.5),
                   fc=(1., 0.8, 0.8),
                   )
         )

ax.text(0, -60, "N", size=25, rotation=0,
         bbox=dict(boxstyle="round",
                   ec=(1., 0.5, 0.5),
                   fc=(1., 0.8, 0.8),
                   )
         )

ax.set_ylabel('Percentage')
ax.set_title('Worktypes group by gender and position')
ax.set_xticks(x)
ax.set_xticklabels([label for label in labels])

ax.legend()
fig.tight_layout()

plt.show()
```

```python
df = pd.read_csv('data/2020/table_15.csv')
df = df.loc[df['Khu vc'] != "Tng s"]
df = df.loc[df['Khu vc'] != "TNG S"]
df = df.loc[df['Khu vc'] != "THNH TH"]
df = df.loc[df['Khu vc'] != "NNG THN"]

df['N'] = df['N'].apply(lambda x : x.replace(" ", ""))
df['Nam'] = df['Nam'].apply(lambda x : x.replace(" ", ""))
df['Chung'] = df['Chung'].apply(lambda x : x.replace(" ", ""))

df['N'] = df['N'].astype('int32')
df['Chung'] = df['Chung'].astype('int32')
df['Nam'] = df['Nam'].astype('int32')
```

```python
data = df.loc[df['Loi'] == "TNG S"]
data = df.groupby(['Khu vc', 'Qu'])['Chung'].sum().reset_index()
fig = px.line(data, x="Qu", y="Chung", color="Khu vc")
fig.show()
```

```python
df = pd.read_csv('data/2020/table_17.csv')
import re
def time2int(time):
    return re.sub("[^0-9]", "", time)
df['Thi gian'] = df['Thi gian'].apply(lambda x: time2int(x))
df = df.loc[df['Khu vc kinh t'] != "Trung bnh"]
```

```python
fig = px.bar(
    df,
    x="Chung",
    y="V tr",
    animation_frame='Thi gian',
#     animation_group="Khu vc kinh t"
#     text="cases",
    color="Khu vc kinh t",
    orientation="h",
)
fig.show()
```

```python
# heat map
corr = df[['Population', 'Tourism', 'Mean_Age', 'Longtitude', 'Latitude', 'Fatality_Rate', 'ConfirmedCases', 'Ah', 'Rh', 'Temp']].corr()

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(8, 6))

sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,mask=mask,cmap='summer_r',vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": 0.5})
```
